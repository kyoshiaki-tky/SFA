```python
def corre_plot_AE(AE_df_train, X_train_pred):
    x = np.linspace(-2,10)
    y = x
    plt.figure(constrained_layout=True)
    plt.title('train correlation')
    plt.scatter(AE_df_train, X_train_pred, alpha=0.02, s=1)
    plt.plot(x,y,color='green', linestyle = '--')
```

```python
def correlation_coefficient(x, y):
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    numerator = np.sum((x - x_mean) * (y - y_mean))
    denominator = np.sqrt(np.sum((x - x_mean) ** 2) * np.sum((y - y_mean) ** 2))
    return numerator / denominator
```

```python
model = load_model("N_NAFLD_NASH_model1_adamax_20dims_linear_liver_data.h5")
model.summary()
```

```python
# Extract the shape of the output from the 'dense_2' layer
intermediate_layer_output_shape = model.get_layer('dense_2').output_shape[1:]

# Define a new input tensor with the same shape as the output of 'dense_21'
new_input = Input(shape=intermediate_layer_output_shape)

# Apply the layers after 'dense_2' to this new input tensor
x = model.get_layer('batch_normalization_2')(new_input)
x = model.get_layer('dense_3')(x)
x = model.get_layer('batch_normalization_3')(x)
x = model.get_layer('dense_4')(x)
x = model.get_layer('batch_normalization_4')(x)
output = model.get_layer('dense_5')(x)

# Create the new model from 'dense_2' to 'dense_5'
decoder_model = Model(inputs=new_input, outputs=output)

# Display the summary of the new model
decoder_model.summary()
```

```python
# decoder data extract
decoder_data_np = decoder_model(df_latent_20dims.values).numpy()
decoder_data_np.shape
```

```python
# calculate gradient means from latent layer to decoder layer at a specific column in latent space
def latent_gradient_mean_each_gene(decoder_model, x, idx_genes, latent_dims):
    x_tensor = tf.convert_to_tensor(x, dtype=tf.float32)
    with tf.GradientTape() as t:
        t.watch(x_tensor)
        loss = decoder_model(x_tensor)
        loss_idx_genes =loss[:,idx_genes:idx_genes+1]
        G = t.gradient(loss_idx_genes, x_tensor).numpy() # shape(102, 20)
    return np.mean(G[:, latent_dims-1])
```

```python
# extract gradients from feature 10 
gradient_means_10 = []
all_genes = 17251
for i in range(all_genes):
    grad_mean_10 = latent_gradient_mean_each_gene(decoder_model, df_latent_20dims, i, 10)
    gradient_means_10.append(grad_mean_10)
    print(i+1,' genes')
    print(grad_mean_10)
```

```python
gradient_means_10_np = np.array(gradient_means_10)
gradient_means_10_np.shape
```
